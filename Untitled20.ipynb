```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.io as pio
pio.renderers.default='notebook'
from ipywidgets import interact,IntSlider
pd.set_option('display.max_columns', None)
```

```python
d = pd.read_excel("/Users/xaviijordi/Downloads/extraccio xavi2.xlsx")
```

```python
d
```

```python
d= d[d['Edad'] != '(nulo)']
```

```python
x = d['Edad'].value_counts().head(10)
```

```python
def plot_bar(sort_by="Index"):
    if sort_by == "Index":
        sorted_counts = x.sort_index()
    elif sort_by == "Frequency Asc":
        sorted_counts = x.sort_values(ascending=True)
    else:  # Frequency Desc
        sorted_counts = x.sort_values(ascending=False)
    
    plt.figure(figsize=(8, 5))
    plt.bar(sorted_counts.index, sorted_counts.values, color='skyblue', alpha=0.7, edgecolor='k')
    plt.title("Interactive Bar Plot")
    plt.xlabel("Category")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
interact(plot_bar, sort_by=["Index", "Frequency Asc", "Frequency Desc"])
```

```python
def gender_barplot(min_age, max_age):
    # Filter data based on age range
    filtered_df = d[(d["Edad"].astype(int) >= min_age) & (d["Edad"].astype(int) <= max_age)]
    
    # Count gender occurrences
    gender_counts = filtered_df["Sexo"].value_counts()
    
    # Create a barplot

    plt.figure(figsize=(8, 5))
    plt.bar(gender_counts.index, gender_counts.values, color='skyblue', alpha=0.7, edgecolor='k')
    plt.title("Gráfico de género en intervalo de edad")
    plt.xlabel("Category")
    plt.ylabel("Cantidad")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

```

```python
interact(
  
    gender_barplot,
    min_age=IntSlider(value=18, min=20, max=60, step=1, description="Edad Mínima"),
    max_age=IntSlider(value=60, min=20, max=60, step=1, description="Edad Máxima")
    
)
```
